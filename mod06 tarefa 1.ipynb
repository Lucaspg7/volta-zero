#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dados_sinasc = pd.read_csv('sinasc_RO_2019.csv')
dados_sinasc.head(5)
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.rc('figure', figsize = (10, 8))
fig, axes = plt.subplots(2, 3)

sns.countplot(ax = axes[0, 0], x='APGAR1', data=dados_sinasc)
sns.countplot(ax = axes[0, 1], x='APGAR5', data=dados_sinasc)
sns.countplot(ax = axes[0, 2], x='ESTCIVMAE', data=dados_sinasc)
sns.countplot(ax = axes[1, 0], x='SEXO', data=dados_sinasc)
sns.countplot(ax = axes[1, 1], x='GESTACAO', data=dados_sinasc)

axes[0, 0].tick_params(axis='x', rotation=45)
axes[0, 1].tick_params(axis='x', rotation=45)
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 1].tick_params(axis='x', rotation=45)

plt.subplots_adjust(wspace=0.35, hspace = 0.35)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sns.barplot(x='CONSULTAS', y='APGAR5', data=dados_sinasc, palette='viridis')
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize = (10, 8))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0, 0], x='SEXO', data=dados_sinasc)
sns.countplot(ax = axes[0, 1], x='GESTACAO', data=dados_sinasc)
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=dados_sinasc)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=dados_sinasc)

axes[0, 0].tick_params(axis='x', rotation=45)
axes[0, 1].tick_params(axis='x', rotation=45)

plt.subplots_adjust(wspace=0.35, hspace = 0.55)
```
