```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
```

```python
arquivo = r"C:\Users\Juliana\Downloads\sinasc_RO_2019.csv"
```

```python
dados_sinasc = pd.read_csv(arquivo)
dados_sinasc.head()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
plt.figure()
sns.boxplot(y="IDADEMAE", data = dados_sinasc)
```

####  2) Faça um boxplot do peso do bebê

```python
plt.figure()
sns.boxplot(y="PESO", data=dados_sinasc)
```

####  3) Faça um histograma da idade da mãe

```python
sns.displot(data = dados_sinasc
           , alpha=.25
           , x='IDADEMAE'
           , bins = 20
           , element = 'step'
           , kde = True
           )
```

#### 4) Faça um histograma do peso do bebê

```python
sns.displot(data = dados_sinasc
           , alpha = 0.75
           , x = 'PESO'
           , bins = 20 
           , element = 'step'
           , kde = True)
```
